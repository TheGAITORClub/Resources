```python
import requests
import os
import pandas as pd
```

```python
def get_yt_thumbnail(id, output_folder='./thumbnails/'):
    thumnail_url = f"http://img.youtube.com/vi/{id}/0.jpg"
    img_data = requests.get(thumnail_url).content
    output_url = os.path.join(output_folder, f"{id}.jpg")
    with open(output_url, 'wb') as handler:
        handler.write(img_data)
    return output_url

get_yt_thumbnail('B0RVWU_nROk')
```

```python
os.listdir('./thumbnails/')
```

```python
data_files = os.listdir('./data/')
data_files
```

```python
videos = pd.DataFrame()

for f in data_files:
    df = pd.read_csv('./data/' + f)
    print(f"Adding file {f}:", len(df), "rows")
    videos = videos.append(df)

videos = videos.drop(columns=['Unnamed: 0'], axis=1)
```

```python
for i, row in videos.iterrows():
    print(f"Getting thumbnail for {row['index']}")
    get_yt_thumbnail(row['index'])
```
